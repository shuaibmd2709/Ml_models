```python
path = '/content/drive/MyDrive/car_insurance_claim.csv'
```

```python
import pandas as pd
import numpy as np
```

```python
raw_data = pd.read_csv(path)
```

```python
pd.set_option('display.max_columns',None)
```

```python
raw_data.head()
```

```python
df1 = raw_data.copy()
```

```python
df1.head()
```

```python
# dropping columns
```

```python
df1.drop_duplicates(inplace=True)
```

```python
df1.drop(columns=['ID','BIRTH','YOJ','PARENT1','MSTATUS'],axis=1,inplace=True)
```

```python
currency_cols = ['INCOME','HOME_VAL','BLUEBOOK','OLDCLAIM','CLM_AMT']

def change_(data,cols):
   for col in cols:
    data[col] = data[col].replace('[\\$,]', '', regex=True).astype('Int64')
   return data
df1 = change_(df1,currency_cols)

```

```python
df1.head()
```

```python

z_cols = ['GENDER','EDUCATION','OCCUPATION','CAR_TYPE']

def change_z(data,cols):
  for col in cols:
    data[col] = data[col].replace('[z_]', '', regex=True)
  return data
df1 = change_z(df1,z_cols)
```

```python
df1['AGE'] = df1['AGE'].fillna(df1['AGE'].mean())
```

```python
df1.info()
```

```python
df1.isnull().sum()
```

```python
binary_cols = ['GENDER','RED_CAR','REVOKED']
mapping = {'yes':1,
       'no':0,
       'M':1,
       'F':0,
       'No':0,
       'Yes':1}

def binary_change(cols,maps,data):
  for col in cols:
    data[col] = data[col].map(maps)
  return data
df1 = binary_change(binary_cols,mapping,df1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df1.copy()
y = df1['CLAIM_FLAG']
```

```python
X.drop(['CLAIM_FLAG'],axis=1, inplace=True)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42,test_size=0.2,)
```

```python
s = (X_train.dtypes == 'object')
object_cols = list(s[s].index)

print("Categorical variables:")
print(object_cols)
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
label_X_train = X_train.copy()
```

```python
ordinal_encoder = OrdinalEncoder()

label_X_train[object_cols] = ordinal_encoder.fit_transform(X_train[object_cols])
```

```python
label_X_train.head()
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy='mean')

imputed_X_train  = pd.DataFrame(imputer.fit_transform(label_X_train))
imputed_X_train.columns = X_train.columns
```

```python
label_X_test = X_test.copy()
label_X_test[object_cols] = ordinal_encoder.transform(X_test[object_cols])
```

```python
imputed_X_test  = pd.DataFrame(imputer.transform(label_X_test))
imputed_X_test.columns = X_test.columns
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import classification_report
```

```python
model = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
model.fit(imputed_X_train, y_train)
```

```python
pred = model.predict(imputed_X_test)
```

```python
model.score(imputed_X_train,y_train)
```

```python
print(classification_report(y_test, pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_2 = RandomForestClassifier(n_estimators=100,max_depth=5,min_samples_split=5, min_samples_leaf=2,)
```

```python
model_2.fit(imputed_X_train,y_train)
```

```python
preds = model_2.predict(imputed_X_test)
```

```python
model_2.score(imputed_X_train,y_train)
```

```python
print(classification_report(y_test, preds))
```
